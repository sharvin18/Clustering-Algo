
 <u>STEPS FOR K-MEDOIDS ALGORITHM:</u><br><br>
 STEP 1: DATASET(D) AND NO OF CLUSTERS(K) WILL BE GIVEN<br>
 STEP 2: CHOOSE K OBJECTS FROM D AS THE INITIAL REPRESENTATIVE OBJECT OR SEEDS<br>
 STEP 3: FIND DISTANCE OF EACH OBJECT IN DATASET FROMS SEEDS<br>
 STEP 4: FIND MIN DISTANCE AND NEREST OBJECT FROM SEEDS<br>
 STEP 5: CALCULATE TOTAL COST<br>
 STEP 6: RANDOMLY SELECT A NON REPRESENTIVE OBJECT AND REPEAT<br>
 STEP 7: COMPUTE SWAPPING COST(S)<br>
 STEP 8: IF S<0 THEN REPEAT<br>
 STEP 9: PRINT CLUSTERS
